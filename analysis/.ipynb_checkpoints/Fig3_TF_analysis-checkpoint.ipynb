# Load data and libs

```python
df_logCPM_all = pd.read_csv("../data/htseq_logCPM_hq.tab.gz", sep="\t", header=0, index_col=0) # CPM

df_libs = pd.read_csv("../data/libs_color3.csv", sep=",", header=0, index_col=0) # sample info
```

# select good ORNs

```python
# Load names of high quality neurons (3/6 markers), cutoff 
names_hq_neurons = []
with open ("../data/names_hq_neurons.txt") as f:
    for line in f:
        names_hq_neurons.append(line.rstrip())
        
# Get names of ORNs
selector0 = (df_libs["label"] == "nSyb_48h_ORN_P0") 
selector1 = (df_libs["label"] == "nSyb_48h_ORN_P1")
selector2 = (df_libs["label"] == "nSyb_48h_ORN_P2")
selector3 = (df_libs["label"] == "nSyb_48h_ORN_P4")
names_ORN = [x for x in list(df_logCPM_all.columns) if ((x in df_libs.loc[selector0].index) 
                      or (x in df_libs.loc[selector1].index) 
                      or (x in df_libs.loc[selector2].index)
                      or (x in df_libs.loc[selector3].index))
                      and x in names_hq_neurons]
print "ORN", len(names_ORN)

# Get names new P4 85A10_ORNs
selector = (df_libs["label"] == "85A10_48h_ORN_P4") 
names_85A10_ORN = [x for x in list(df_logCPM_all.columns) if ((x in df_libs.loc[selector].index) and x in names_hq_neurons)]
print "New_85A10_ORN_P4", len(names_85A10_ORN)

# Get names AM29_ORNs
selector = (df_libs["label"] == "Am29_48h_ORN_P4") 
names_AM29_ORN = [x for x in list(df_logCPM_all.columns) if ((x in df_libs.loc[selector].index) and x in names_hq_neurons)]
print "AM29_ORN", len(names_AM29_ORN)
```

```python
# Filter for these cells
names_All_ORN = names_ORN + names_85A10_ORN +names_AM29_ORN
df_All_ORN = df_logCPM_all[list(names_All_ORN)]

df_AM29_ORN = df_logCPM_all[list(names_AM29_ORN)]
df_85A10_ORN = df_logCPM_all[list(names_85A10_ORN)]


print "ORN_All:", df_All_ORN.shape[1]
print "AM29_ORN:", df_AM29_ORN.shape[1]
print "85A10_ORN:", df_85A10_ORN.shape[1]

# Display number of cells of each genotype to confirm
#df_libs.loc[df.columns]["genotype"].value_counts()
```

# Load ICIM genes and plot tSNE 

```python
X = df_All_ORN
X.shape
```

```python
names_ORN_ICIM = []
with open("../data/ICIM_ORN_408_Genes.txt") as f:
    for line in f:
        names_ORN_ICIM.append(line.rstrip())
print "names_ORN_ICIM", len(names_ORN_ICIM)
```

```python
X.shape
Z = X.loc[names_ORN_ICIM]
Z.shape
```

```python
df_libs = pd.read_csv("../data/libs_color3_PN_ORN.csv", sep=",", header=0, index_col=0) # sample info
```

```python
# Calculate TSNE
reload(sct)
from sklearn.manifold import TSNE

for i in range(18,19): # PC = i
    myPCA = sct.PCA(Z, X, i) # number of significant PCs was determined using shuffling performed elsewhere in parallel fashion
    myPCA.pca()
    Z_pca = pd.DataFrame(myPCA.X_pca.T)

    for j in range(15,16): # perplexity = j
        reload(sct)
        from sklearn.manifold import TSNE
        myTSNE = sct.TSNE(Z_pca, X, df_libs)
        
        myTSNE.calc_TSNE(perplexity=j, learning_rate=500)
        fig, ax = plt.subplots(1, 1, figsize=(6,6))
        myTSNE.plot(fig, ax, colorMode="genotype")
        ax.set_title(str(i)+"PCs_" + str(j) +"perplexity_lr500" , y=1.02)
        #fig.savefig("../Figures/Fig_1/Fig1_ICIM_ORN_408Genes_" + str(i) +"_PCs_" + str(j) + "_perplexity" + "_lr500.pdf")
```

# HDBSCAN to call clusters

```python
import hdbscan
clusterer = hdbscan.HDBSCAN(min_cluster_size=6, min_samples=6).fit(myTSNE.X_tsne)
labels_HDBSCAN = clusterer.labels_
print "Clusters", max(labels_HDBSCAN)+1
```

```python
# Plot TSNE colored by cluster assignment with labels
fig, ax = plt.subplots(1, 1, figsize=(6,6))
myTSNE.plot(fig, ax, colorMode="custom", colorBy=labels_HDBSCAN, cmap="nipy_spectral")
    
#fig.savefig("../Figure/Fig_1/Fig1_ICIM_ORN_408Genes_" + "HDSCAN.pdf")
```

```python
# Plot TSNE colored by cluster assignment with labels
fig, ax = plt.subplots(1, 1, figsize=(6,6))
myTSNE.plot(fig, ax, colorMode="custom", colorBy=labels_HDBSCAN, cmap="nipy_spectral")
    
for label in pd.Series(labels_HDBSCAN).value_counts().index:
    if label == -1: continue
    x_bar = np.mean(myTSNE.X_tsne[labels_HDBSCAN == label, 0])
    y_bar = np.mean(myTSNE.X_tsne[labels_HDBSCAN == label, 1])
    ax.text(x_bar, y_bar, label, size=10, color = 'r')
    
```

# Use different threshold for positive cluster definition

```python
# this value will be used to determine when to call a TF is positive for one cluster
Expr_Fraction_value = 0.3
```

```python
TFs = df_TFs.index
print "detected TFs", len(TFs)
```

# calculate TF expression fraction and mean value in positive clusters

```python
# create individual CSV files with cels for each cluster
for label in df_labels["label"].value_counts().index:
    if label == -1: continue
    names1 = list(df_labels[df_labels["label"] == label].index)
    frac_expr = np.sum(df[names1].loc[TFs] > 3, axis=1) / float(len(names1))
    Mean_Expr = df[names1].loc[TFs].mean(axis=1)
    df_TF_label = pd.DataFrame({"TFs": TFs, "Expr_Fraction_"+str(label): frac_expr, "Mean_Expr_"+str(label): Mean_Expr,})
    df_TF_label.set_index("TFs", inplace=True)
    outfile = "Fig3_cluster_TFs/TFs_in_cluster_" + str(label) + ".csv"
    df_TF_label.to_csv(outfile)
```

```python
# create individual CSV files with cels for each cluster
Mydf = {}
for label in df_labels["label"].value_counts().index:
    if label == -1: continue
    infile = "Fig3_cluster_TFs/TFs_in_cluster_" + str(label) + ".csv"
    df_TF_combine = pd.read_csv(infile, index_col=0, header=0)
    Mydf[label] = df_TF_combine["Expr_Fraction_"+str(label)]
    #Mydf[label] = df_TF_combine
```

```python
# combine all Fraction dataframe to a single one
Mydf_combined_fraction = pd.concat([Mydf[0], Mydf[1],Mydf[2], Mydf[3],Mydf[4], Mydf[5],Mydf[6], Mydf[7],Mydf[8], 
                           Mydf[9], Mydf[10],Mydf[11], Mydf[12],Mydf[13], Mydf[14],Mydf[15], Mydf[16],Mydf[17], 
                           Mydf[18], Mydf[19],Mydf[20], Mydf[21],Mydf[22], Mydf[23],Mydf[24], Mydf[25],Mydf[26], 
                           Mydf[27], Mydf[28],Mydf[29], Mydf[30],Mydf[31], Mydf[32]], axis=1)
Mydf_combined_fraction.head(35)
```

```python
Mydf_combined_fraction[ Mydf_combined_fraction >= Expr_Fraction_value] = 1
Mydf_combined_fraction[ Mydf_combined_fraction < Expr_Fraction_value] = 0
#Mydf_combined_fraction.tail(35)
```

# Expression level and dropout rate: all genes,  TFs and CSMs

```python
df = df_All_ORN
df.shape
```

```python
# set read number threshold for dropout
dropout_threshold = 3
```

```python
# What is the dropout rate of the marker genes used to filter cells?

genes = TFs

dropout_rate = np.sum(df >= dropout_threshold, axis=1) / df.shape[1]
dropout_rate = np.sum(df >= dropout_threshold, axis=1)



fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.set_xlim(1, 15)

x = np.mean(df, axis=1)
y = dropout_rate
ax.scatter(x, y, marker=".",linewidths=0.2, c = 'white')

x = np.mean(df, axis=1).loc[genes]
y = dropout_rate.loc[genes]
ax.scatter(x, y, marker=".", c="r", s=30)

ax.set_xlim(left=0)
ax.set_ylim(0,1100)
ax.set_xlabel("Mean expression level (log2(CPM+1))")
ax.set_ylabel("Fraction of cells in which gene\nis detected (log2(CPM+1) >= 3)")
  
plt.tight_layout()

fig.savefig("../Figure/Fig_3/Fig3_AllGenes_TFs_detected_fraction.pdf")
```
